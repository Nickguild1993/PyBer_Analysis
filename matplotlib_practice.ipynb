```python
#import dependencies 

import matplotlib.pyplot as plt

```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with an ax.plot()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Creat the plot and add a label for the legnd.

plt.plot(x_axis, y_axis, label = "Boston")

# Create labels for the x and y axes

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0,45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Creat the plot and add a label for the legnd.

plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")

# Create labels for the x and y axes

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0,45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add a grid.

plt.grid()

# Add the legend.

plt.legend()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, marker = "^", color = "green", linewidth = 2, label = "Boston")

# Create labels for the x and y axes

ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.

ax.set_ylim(0,45)

# Create a title.

ax.set_title("PyBer Fare by Month")

# Add a grid.

ax.grid()

# Add the legend.

ax.legend()
```

```python
# Making a bubble chart

plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data * 3)
```

```python
# Skill Challenge 

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, c = "blue", s = [i * 5 for i in y_axis], label = "Boston")
ax.set_xlabel("Month")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fares by Month")
ax.legend()
```
